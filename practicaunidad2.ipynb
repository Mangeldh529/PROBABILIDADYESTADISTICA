```python
import math as mt
from itertools import combinations
from itertools import combinations_with_replacement
import numpy as np
```

```python
nombres = ["Miguel", "Angel", "Alejandro", "Adrian Marcelo", "Moises", "Email", "Mario", "Christian", "Angelico", "Ever"]
n = len(nombres)
r = 2

Cn = mt.factorial(n) / (mt.factorial(r) * mt.factorial(n-r))

print("Existen", Cn , " Posibles combinaciones sin repeticion del total de",
     n, " nombres", " en grupos de ", r , "en", r)

print("Aqui las combinaciones")
combinaciones_nombres = list(combinations(nombres, r))


for combinacion in combinaciones_nombres:
    print(combinacion)
```

```python
asistentes = ["a1", "b2", "c3", "d4", "e5", "f6", "g7", "h8", "i9", "j10"]
n = len(asistentes)
k = 4

Pr = mt.factorial(n) / (mt.factorial(n-k))

num_permutations = len(list(permutations(range(n), k)))
print("numero de permutaciones:", num_permutations)

print("Existen", Pr , " posibles permutaciones con repeticion del total de ",
     n, " asistentes ", " en grupos de ", k ," en ", k)

print("Aqui las permutaciones")
permutaciones_asistentes = list(permutations(asistentes, k))

for permutacion in permutaciones_asistentes:
    print(permutacion)
```
